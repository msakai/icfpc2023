```python
from pathlib import Path
import json
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
def plot_problem(ax, problem):
    ax.set_xlim((0, problem["room_width"]))
    ax.set_ylim((0, problem["room_height"]))

    r = patches.Rectangle(
        xy=problem["stage_bottom_left"],
        width=problem["stage_width"],
        height=problem["stage_height"],
        ec='black', fc='black', fill=True
    )
    ax.add_patch(r)

    xs = []
    ys = []
    for attendee in problem["attendees"]:
        xs.append(attendee["x"])
        ys.append(attendee["y"])    
    plt.scatter(xs, ys)

    for pillar in problem["pillars"]:
        c = patches.Circle(xy=pillar["center"], radius=pillar["radius"], fc='g', ec='g')
        ax.add_patch(c)
```

```python
for i in range(56, 90+1):
    with open(f"../problems/{i:03}.json") as f:
        problem = json.load(f)
    fig = plt.figure()
    ax = plt.axes()
    plot_problem(ax, problem)
    plt.title(f"problem {i}")
    plt.show()
```
